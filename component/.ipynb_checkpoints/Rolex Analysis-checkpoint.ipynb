```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("/Users/junyuwu/Downloads/rolex_scaper_clean.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
model_count = df["model"].value_counts().nlargest(10)
model_count
```

```python
mylabels = model_count.index
mycolors = ["yellow", "c", "b", "#4CAF50","red"]
plt.figure(figsize=(10,10))
plt.pie(model_count,labels = mylabels, autopct="%.1f%%", colors = mycolors, textprops={"size" : "x-large"})
plt.title("Model Proportion", color="blue", fontsize=35,loc='center')
plt.legend(bbox_to_anchor=(-0.5, 1), loc=2, borderaxespad=0.);
```

```python
model_byprice = df.groupby("model")["price"].sum().nlargest(10)
model_byprice
```

```python
mylabels = model_byprice.index
mycolors = ["yellow", "c", "b", "#4CAF50","red"]
plt.figure(figsize=(10,10))
plt.pie(model_byprice,labels = mylabels, autopct="%.1f%%", colors = mycolors, textprops={"size" : "x-large"})
plt.title("Model Total Price Proportion", color="blue", fontsize=35,loc='center')
plt.legend(bbox_to_anchor=(-0.5, 1), loc=2, borderaxespad=0.);
```

```python
bymovement = df.groupby("movement")["price"].mean()
bymovement
```

```python
x_axis = bymovement.index
y_axis = bymovement
plt.figure(figsize=(12, 7))
plt.style.use('seaborn')
plt.bar(x_axis, y_axis, width=0.5, color="blue")
plt.title('Movement Average Price',fontsize=16)
plt.xlabel('Movement Categories',fontsize=16)
plt.ylabel('Average Price',fontsize=16);
#Automatic or Manual winding
```

```python
by_material = df.groupby("case material")["price"].mean()
by_material
```

```python
x_axis = by_material.index
y_axis = by_material
plt.figure(figsize=(12, 7))
plt.style.use('seaborn')
plt.bar(x_axis, y_axis, width=0.5, color="blue")
plt.title('Case Material Average Price',fontsize=16)
plt.xlabel('Case Material',fontsize=16)
plt.ylabel('Average Price',fontsize=16);
#Platinum
```
